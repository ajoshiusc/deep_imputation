```python
import matplotlib.pyplot as plt
import numpy as np

# Data and labels for the first set of bars

values = [
    [97.4, 100, 99.1], 
    [100, 99.4, 98.2], 
    [99.5, 91.9, 80.1], 
    [99.1, 98.9, 96.3],
]
labels = ['WT', 'TC', 'ET']
colors = ['#7A0B6F', '#E02C24', '#FCFDBF']

x_vals = [np.arange(len(values[0]))]
for i in range(1, len(values)):
    x_vals.append(x_vals[i-1] + len(values[0]) + 1.5)

for i in range(len(values)):
    plt.bar(x_vals[i], values[i], color=colors, width=1.0)

# Adding titles and labels
plt.title('Drop one modality and train with the rest')
# plt.xlabel('Categories')
plt.ylabel('Relative DICE')

# Set y-axis limit
plt.ylim([70, None])

# Set x-axis ticks to match the labels and adjust tick positions
plt.xticks(np.concatenate(x_vals), labels*len(values))

# Add labels below each set
for i in range(len(values)):
    plt.text(np.mean(x_vals[i]), 55, 'Set-1', ha='center', va='center', fontsize=12)

# Show the plot
plt.show()
```
